```python
import numpy as np
import pandas as pd
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt

%matplotlib inline

from IPython.display import Image
```

```python
minedata = pd.read_csv("C:\\Users\\goelp\\OneDrive\\Documents\\Python Projects\\Mine vs rock classification - Logistic regression\\Copy of sonar data.csv", header =None)
minedata.head()
```

```python
minedata.describe()
```

```python
minedata.dtypes
```

```python
minedata[60].value_counts()
```

```python
minedata.groupby(60).mean()
```

```python
# Seperating the Data and Labels
#Supervised Learning Model

#Labels are Rock and Mine

X = minedata.drop(columns = 60, axis=1)
Y = minedata[60]
```

```python
X.head()
```

```python
#Splitting the data into training and Test data
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.1, stratify = Y, random_state = 1)
```

```python
print(X.shape, X_test.shape, X_train.shape)
```

We will be using LOGISTIC REGRESSION MODEL

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
#Model Evaluation
```

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
training_data_accuracy*100
```

```python
confusion_matrix(X_train_prediction, Y_train)
```

```python
#Accuracy on Test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
test_data_accuracy*100
```

```python
confusion_matrix(X_test_prediction, Y_test)
```
